## Sub PLots:
A subplot in Matplotlib is a way to place multiple plots (charts) in a single figure. Instead of creating separate figures subplots let us organize plots in rows and columns within the same window.


in this subplot we compare countries gdp per capita and their generosity rates.


in first graph the too 10 most happy countries generosity is shown and


in second one gdp per capita of that countries are shown in the form of graph.



```python
fig, ax = plt.subplots(figsize=(30,12))

x=df['Country or region'][0:6]
y=df['Generosity'][0:6]
y1=df['GDP per capita'][0:6]
plt.subplot(1,2,1)
plt.bar(x,y)
plt.xticks(fontsize=22)
plt.yticks(fontsize=20)
plt.title('Generosities by countries',fontsize=40)
plt.xlabel('Country',fontsize=25)
plt.ylabel('Score',fontsize=25)

plt.subplot(1,2,2)
plt.bar(x,y1)
plt.xticks(fontsize=22)
plt.yticks(fontsize=20)
plt.title("GDP per capita by country",fontsize=40)
plt.xlabel('Country',fontsize=22)
plt.ylabel('Score',fontsize=25)
plt.show()

```

## Sub bar plot by using seaborn:
in this subplot we are going to visualize the perception of corruption and freedom of choices of top 10 world most happy countries.

in first graph we are visualizing freedom of choices and in second one we are going to visualize perception of corruption in top 10 most happy countries in world according to data set.



```python

fig, ax = plt.subplots(1, 2,figsize=(30,12))

data = df[0:10]

sns.set(font_scale = 3)
ax1 = sns.barplot(data= data, x= 'Country or region',y = 'Freedom to make life choices',palette='viridis', ax = ax[0])
ax1.set_xticks(ax1.get_xticks())
ax1.set_xticklabels(data['Country or region'], rotation=60,fontsize = 17)
ax1.set_xlabel('Country',fontsize=25)
ax1.set_ylabel('Freedom of choices',fontsize=25)
ax1.set_title('Freedom of choice in 10 most happy countries')


ax2 = sns.barplot(data= data, x= 'Country or region',y = 'Perceptions of corruption',palette='viridis', ax=ax[1])
plt.xticks(rotation=60,fontsize = 17)
ax2.set_xlabel('Country',fontsize=25)
ax2.set_ylabel('Perception of corruption',fontsize=25)
ax2.set_title('perception of corruption in top 10 countries')

plt.show()
```

##GG plot:
In Python, plotnine is a library that brings ggplot-like syntax to Python.

GGplot is a popular data visualization system originally from R, based on the Grammar of Graphics. It provides a structured way to create complex and elegant plots by combining layers of data, aesthetics, and geometric objects.

Key points:

Data  the dataset you want to plot.

Aesthetics (aes)  defines what goes on the axes, color, size, etc.

Geoms (geom_)  the type of plot, like geom_bar() for bar charts or geom_line() for line charts.

Facets  for creating subplots based on a variable.

Themes & labels  customize titles, axes, and overall appearance.

Plot a bar chart using GGPLOT :


this code cretes a bar chart for least 10 countries and their happiness score according to data set iam using.

```python

chart_data = df.nsmallest(10, 'Score')


(
    ggplot(chart_data, aes(x='Country or region', y='Score'))
    + geom_col(fill="gray")
    + labs(title='last 10 Countries by Happiness Score',
           x='Country', y='Happiness Score')
    + theme(figure_size=(10,6)))

```

## Cross Tabulation table:
A cross-tabulation table (or crosstab) is a way to summarize the relationship between two categorical variables in a table format.


in this tabel first i categorize the happiness score and then made a cross tabulation tabel to make it easy to understand.




```python

pd.crosstab(df['Country or region'], pd.cut(df['Score'], bins=[0,5,6.5,8], labels=['Low','Medium','High']))

```

## Heat Map of correlation in world happiness data:

This heatmap shows how different factors are related to happiness in different countries.
We see that happiness scores are strongly linked to GDP, life expectancy, and social support meaning richer countries where people live longer and have stronger communities tend to be happier.
On the other hand factors like generosity dont show much connection with happiness.


the red boxes shows positive correlation while blue boxes negative correlation between diffrent factors.


```python
numeric_df = df.select_dtypes(include="number")
corr = numeric_df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap of World Happiness data")
plt.show()
```
