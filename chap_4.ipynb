## Histogram:
A histogram in Seaborn is a bar-like plot that shows how often different ranges of values occur in a dataset.

in this code i make a istogram of sepal length of the data that is given in iris data set.

```python
plt.figure(figsize=(12,6))
ax=sns.histplot(df['sepal_length'],kde=True,bins=10)
ax.set_xlabel('sepal_length')
ax.set_ylabel('Frequency')
ax.set_title('Frequency distribution of sepal length')
plt.show()
```

## Box plot:
A boxplot finds and shows the summary of a datasets spread (minimum, maximum, median, quartiles, and outliers) in one compact graph.

in this code i make a box plot that simply shows how the sepal width measurements are spread out.

The red box shows where most of the values (the middle half) fall, centered around a typical value of 3.0.

The lines and circles show that some flowers have much smaller (around 2.0) or much larger (above 4.0) sepal widths than the others.



```python
plt.figure(figsize=(12,6))
ax=sns.boxplot(df['sepal_width'],color='r')
ax.set_ylabel('sepal_width')
ax.set_title("Distribution of sepal width")
plt.show()
```

## violin plot:
t is just similar to boxplot but we can say it is more fansy and easy to visualize

I create a violin plot of the petal length column from my dataset.

A violin plot combines a boxplot with a density curve, so it not only shows the median and quartiles (like a boxplot) but also the shape of the data distribution (where values are more frequent).

```python
plt.figure(figsize=(12,6))
ax=sns.violinplot(df['petal_length'],color='g')
ax.set_ylabel('petal_length')
ax.set_title("Distribution of petal length")
plt.show()
```

## Bar Plot using countplot method in seaborn:
in this code i make a bar graph by using seaborn library that shows each species have equal number of samples in respective data set.


```python
sns.countplot(data=df,x='species',color='g')
plt.title('Count of each species')
plt.show()
```

## Pie chart:
a pie chart is construct by using this code that shows part of each species in sample data set and after analyzing as we can see all species have equal percentage in given data set.


```python
counts = df['species'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title("Distribution of Iris Species")
plt.show()
```
