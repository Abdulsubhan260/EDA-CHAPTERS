##Quartile:
A quartile divides a dataset into four equal parts (quarters) after sorting the values in order.

Q1: 1st quartile / 25th percentile

Q2: 2nd quartile / 50th percentile / Median

Q3: 3rd quartile / 75th percentile Percentile

##Percentile:
A percentile tells us the value below which a given percentage of data falls.


in my case iam going to find 25th,50th and 75th quartile of math score.






25th percentile (Q1) = 57


it means 25% of students scored below 57 in math and 75% score more than 57


50th percentile (Q2 / Median) = 66



it means 50% of students scored below 66.


75th percentile (Q3) = 77




it means  75% of students scored below 77.


```python
percentile = np.percentile(df['math score'],[25,50,75])
percentile
```

InterQuartile Range:
The Interquartile Range (IQR) is the range between the 1st quartile (Q1) and the 3rd quartile (Q3).

𝐼𝑄𝑅 𝑄 3 − 𝑄 1 : IQR = Q3−Q1

Why i use it: As i take "math score" from my data set so in my data set is shows the range between the


 Q1: (25th percentile) in which the math score is equal to 57

And

Q3 :(75th percentile)in which the math score is equal to 77

So:

IQR = 77-57

so,interquartile range in case of math score is 20

```python
IQR = stats.iqr(df["math score"], interpolation = 'midpoint')
IQR
```

## Box Plot:

Median scores (the middle line in each box) are higher for the standard lunch group.

Students with standard lunch mostly scored higher and closer together, while students with free/reduced lunch had lower average scores and more very low marks.


so in summary:


in this data set the box plot shows both in case of male and female the students with standard lunch get more score in maths as compare to the students who have free/reduced lunch.

```python
sns.boxplot(x="gender",y="math score",hue="lunch",data=df)
plt.suptitle("box plot")
plt.show()
```

## Heat Map:


in this heat map i compared all(math score,reading score and writting scores) and analyse correlation between them and i found in most of cases they all have strong positive correlation with each other means a student who scored high marks in math have very much chances that particular student also scored good in reading and writting.

```python
sns.heatmap(df.select_dtypes(['number']).corr(),annot=True,cmap="icefire")
plt.show()
```

## Histogram:

according to data set iam using i make this histogram to see that lunch type has any effect on marks of students or not and then i analyse that it has big effect on students obtained marks.students with standard lunch both in male and female score high as compare to students who have free/reduced lunch.

```python
sns.histplot(data=df, x="math score", kde=True, bins=10, hue="lunch", element="step")
plt.title("Math Score Distribution by Lunch Type")
plt.show()
```
